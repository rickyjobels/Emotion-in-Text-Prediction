### Import WordNetLemmatizer, stopwords

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import re
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
```

### Create a function to preprocess the text (lemmatize,lower,remove stopwords, remove special characters)

```python
def emotions(emotion):
    emotion=emotion.lower()
    emotion = re.sub(r"[^0-9a-zA-Z\s]", '', emotion)
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    words = nltk.word_tokenize(emotion)
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return ' '.join(words)
```

### Apply the function to Text in our dataframe

```python
df['Text'] = df['Text'].apply(emotions)
```

```python
df['Text']
```

### View some of the texts after preprocessing

```python
for i in df['Text'].head(10):
    print(i)
    print()
```

### Convert text to vectors

```python
X=df['Text']
y=df['Emotion']
```

```python
vectorise = TfidfVectorizer()
```

```python
X = vectorise.fit_transform(X)
```

```python
X.shape
```

### Split the dataset into training and Testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=100)
```

### Create a Random forest classifier model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random = RandomForestClassifier(n_estimators=50)
```

### Train the model

```python
random.fit(X_train,y_train)
```

### Check the score of the model

```python
random.score(X_train,y_train)
```

### Make predictions with X_test

```python
y_pred = random.predict(X_test)
```

### Check the accuracy of our prediction

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test,y_pred)
```

### Create confusion matrix

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt='d')
```

### Create classification report

```python
print(classification_report(y_test,y_pred))
```
